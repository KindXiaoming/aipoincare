```python
import numpy as np
import matplotlib.pyplot as plt

nx, ny = (200, 200)
x = np.linspace(0, 2000, nx)
y = np.linspace(0, 0.002, ny)
xv, yv = np.meshgrid(x, y)
c = xv * yv > 1
```

```python
plt.figure(figsize=(5,5))
plt.scatter(xv,yv,c=c,s=10)
plt.xlim(0,2000)
plt.ylim(0,0.002)
plt.ylabel(r"$\epsilon$",fontsize=30)
plt.xticks([0,0.5*1000,1000,1.5*1000,2*1000],fontsize=15)
plt.yticks([0,0.5*0.001,0.001,1.5*0.001,2*0.001],fontsize=15)
plt.xlabel("n(orbits)",fontsize=30)
plt.text(0.8*1000,1.5*0.001,r"$N_{eff}(C)=2$",color="red",fontsize=30)
plt.text(0.2*1000,0.4*0.001,r"$N_{eff}(C)=3$",color="red",fontsize=30)
plt.title("Kepler Problem",fontsize=25)
```

```python
nx, ny = (200, 200)
x = np.linspace(0, 2, nx)
y = np.linspace(0, 2, ny)
xv, yv = np.meshgrid(x, y)
c = (xv**2) * yv > 1

plt.figure(figsize=(5,5))
plt.scatter(xv,yv,c=c,s=10)
plt.xlim(0,2)
plt.ylim(0,2)
plt.ylabel("k",fontsize=30)
plt.xticks([0,0.5,1,1.5,2],fontsize=15)
plt.yticks([0,0.5,1,1.5,2],fontsize=15)
plt.xlabel(r"$A_{\rho}$",fontsize=30)
plt.text(0.85,1.5,r"$N_{eff}(C)=1$",color="red",fontsize=30)
plt.text(0.2,0.4,r"$N_{eff}(C)=2$",color="red",fontsize=30)
plt.title("Magnetic Mirror",fontsize=25)
```

```python
nx, ny = (200, 200)
x = np.linspace(0, 2, nx)
y = np.linspace(0, 2, ny)
xv, yv = np.meshgrid(x, y)
c = (xv**2) * yv > 1

plt.figure(figsize=(5,5))
plt.scatter(xv,yv,c=c,s=10)
plt.xlim(0,2)
plt.ylim(0,2)
plt.ylabel(r"$k(m,l,g)$",fontsize=20)
plt.xticks([0,0.5,1,1.5,2],fontsize=15)
plt.yticks([0,0.5,1,1.5,2],fontsize=15)
plt.xlabel(r"$\theta_0$",fontsize=30)
plt.text(0.85,1.5,r"$N_{eff}(C)=1$",color="red",fontsize=30)
plt.text(0.2,0.4,r"$N_{eff}(C)=2$",color="red",fontsize=30)
plt.title("Double Pendulum",fontsize=25)
```
